```python
# b)
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the first image into a NumPy array
img1 = np.array(Image.open('sample_one.jpg'))

# Flip the first image sideways
flipped_img1 = np.fliplr(img1)

# Load the second image into a NumPy array
img2 = np.array(Image.open('sample_two.png'))

# Flip the second image sideways
flipped_img2 = np.fliplr(img2)

# Display the original and flipped images for both images side by side
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
ax1.imshow(img1)
ax1.set_title('Original Image 1')
ax2.imshow(flipped_img1)
ax2.set_title('Flipped Image 1')
ax3.imshow(img2)
ax3.set_title('Original Image 2')
ax4.imshow(flipped_img2)
ax4.set_title('Flipped Image 2')

plt.show()
```
