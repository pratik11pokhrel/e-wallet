```python
# c)
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the first image into a NumPy array and display its green channel
img1 = np.array(Image.open('sample_one.jpg'))
green_img1 = np.zeros_like(img1)
green_img1[:,:,1] = img1[:,:,1]
plt.imshow(green_img1)
plt.show()

# Load the second image into a NumPy array and display its green channel
img2 = np.array(Image.open('sample_two.png'))
green_img2 = np.zeros_like(img2)
green_img2[:,:,1] = img2[:,:,1]
plt.imshow(green_img2)
plt.show()
```
