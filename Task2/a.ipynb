```python
# a)
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the first image into a NumPy array
img1 = np.array(Image.open('sample_one.jpg'))

# Split the first image vertically in half
height1, width1, channels1 = img1.shape
split_point1 = width1 // 2
left_half1 = img1[:, :split_point1, :]
right_half1 = img1[:, split_point1:, :]

# Display the first image and the two halves side by side
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax1.imshow(img1)
ax1.set_title('Original Image 1')
ax2.imshow(left_half1)
ax2.set_title('Left Half 1')
ax3.imshow(right_half1)
ax3.set_title('Right Half 1')
plt.show()

# Load the second image into a NumPy array
img2 = np.array(Image.open('sample_two.png'))

# Split the second image vertically in half
height2, width2, channels2 = img2.shape
split_point2 = width2 // 2
left_half2 = img2[:, :split_point2, :]
right_half2 = img2[:, split_point2:, :]

# Display the second image and the two halves side by side
fig, (ax4, ax5, ax6) = plt.subplots(1, 3)
ax4.imshow(img2)
ax4.set_title('Original Image 2')
ax5.imshow(left_half2)
ax5.set_title('Left Half 2')
ax6.imshow(right_half2)
ax6.set_title('Right Half 2')
plt.show()
```
